# Predicciones de Sentiment

Este programa carga un csv con tweets y en base a ellos da una prediccion de la positividad, negatividad, y neutralidad de cada uno.

Luego, exporta estos datos a un archivo .csv llamado "results.csv"

Se utiliza un modelo de Hugging Face para las predicciones

Modelo de Hugging Face: https://huggingface.co/lxyuan/distilbert-base-multilingual-cased-sentiments-student?

```python
from transformers import pipeline
import pandas as pd
```

```python
# Definimos el modelo de clasificación
sentiment_classifier = pipeline(
    model="lxyuan/distilbert-base-multilingual-cased-sentiments-student",
    return_all_scores=True
)
```

```python
# Cargamos los datos a una DataFrame de Pandas
df = pd.read_csv('data.csv')
df.head()
```

```python
# calculamos los sentiments y lo agregamos al DataFrame
def get_sentiments(df):
  for i in df.index:
    print("processing row ", i)
    results = sentiment_classifier (df.loc[i, "tweet"])
    results = results[0]

    # add results to dataframe
    for j in range(len(results)):
      df.loc[i, results[j]["label"]] = results[j]["score"]
```

```python
# Ejecutamos la funcion para calcular los sentiments e imprimimos los primeros 5 renglones
get_sentiments(df)
df.head()

# Cargamos los resultados a un archivo llamado results.csv
df.to_csv('results.csv')
```
